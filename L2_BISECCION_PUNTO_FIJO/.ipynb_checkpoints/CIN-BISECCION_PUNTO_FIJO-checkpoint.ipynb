<h3>Representaciones gráficas en el plano</h3>

Veamos algunas opciones para representar gráficamente una función $f$, así como algunas expresiones para la representación de rectas (segmentos) y puntos del plano.

```python
x = var('x')

f(x) = 3*sin(x/2) - 5*cos(x)  ### Observación: la función seno se expresa como sin()

show(f(x))

plot(f(x), (x, 0, 10), color = 'blue')
```

<p>Podemos utilizar ciertos patrones para el trazado de la línea en las gráficas con la opción <span style="color: #0000ff;">linestyle='patron'</span>, donde el patrón puede tomar ciertos valores como <span style="color: #0000ff;">linestyle=':'</span>  para línea discontinua de puntos o <span style="color: #0000ff;">linestyle='-.'</span> para linea discontinua que alterna segmentos y puntos. Puedes obtener más información sobre opciones gráficas evaluando la orden <span style="color: #0000ff;"> plot?</span></p>

```python
plot(f(x), (x, 0, 10), color = 'green', linestyle = ':')
```

```python
line([(0, 0), (6, 6)], color = 'red', linestyle = '-.' )
```

<p>Recuerda que los gráficos se pueden "sumar". Podemos representar varios gráficos simultáneamente con la función <span style="color: #0000ff;">show()</span>:</p>

```python
show(
    plot(f(x), (x, 0, 10), color = 'blue') +
    line([(0, 0), (6, 6)], color = 'red') +
    point([(3, 4)], color = 'green', size = 50)
)
```

<p>Incluso podemos determinar una ventana o rango de valores para seleccionar un área de la gráfica con los valores opcionales <span style="color: #0000ff;">xmin, xmax, ymin,ymax</span>:</p>

```python
show(
    plot(f(x), (x, 0, 10), color = 'blue')
    + line([(0, 0), (6, 6)], color = 'red')
    + point([(3, 4)], color = 'green', size = 50),
    xmin = 0, xmax = 4, ymin = 0, ymax = 6)
```

<p>También podemos crear un marco de líneas discontinuas que centre nuestra atención en un área determinada del gráfico:</p>

```python
show(
    plot(f(x), (x, 0, 10), color = 'blue') +
    line([(1, 1), (4, 1)], linestyle = ':', color = 'red' )
    + line([(4, 1), (4, 6)], linestyle = ':', color = 'red' )
    + line([(4, 6), (1, 6)], linestyle = ':', color = 'red' )
    + line([(1, 6), (1, 1)], linestyle = ':', color = 'red' ))
```

<p>Vamos a crear una función que genere  un rectángulo como el anterior, el cual vendrá determinado por dos puntos diagonalmente opuestos, desde esquina inferior izquierda a esquina superior derecha:</p>

```python
def rectangulo(vertice1, vertice2, color = 'red', estilo = ':'):
    (xmin, ymin) = vertice1
    (xmax, ymax) = vertice2
    return (line([(xmin, ymin), (xmax, ymin)], linestyle = estilo, color = color) 
            + line([(xmax, ymin), (xmax, ymax)], linestyle = estilo, color = color) 
            + line([(xmax, ymax), (xmin, ymax)], linestyle = estilo, color = color) 
            + line([(xmin, ymax), (xmin, ymin)], linestyle = estilo, color = color))
```

```python
show(
    plot(f(x), (x, 0, 10), color = 'blue')
    + rectangulo((4, -2), (6, 2))
    + rectangulo((1, 2), (3, 6), color = 'green')
)
```

<p><a name="biseccion"></a></p>
<h3>Método de  Bisección</h3>
<p>Recordemos que el método de bisección pretende resolver ecuaciones de la forma $$f(x)=0.$$</p>
<p>Ahora resolveremos la ecuación no lineal $f(x)=0$ del ejemplo anterior usando el método de  bisección:</p>
<p>$$3\sin\left(\frac{x}{2}\right)-5\cos(x)=0$$</p>
<p>Dibujamos su gráfica entre 0 y 10 para localizar los puntos de corte con el eje OX que son las soluciones de nuestra ecuación.</p>

```python
plot(f(x), (x, 0, 10))
```

<p>Esta ecuación tiene infinitas raíces, debido a la periodicidad de las funciones sen y cos. Vamos a calcular la menor raíz positiva.</p>
<p>Si consideramos el intervalo [0,2], tenemos que se verifican las hipótesis del Teorema de Bolzano pues la función cambia de signo en los extremos del intervalo, esto es   $f(0)\cdot f(2)<0$:</p>

```python
n(f(0) * f(2)) < 0
```

<p>Aunque la representación gráfica no deja lugar a dudas sobre la existencia de una única raíz en el intervalo [0,2], vamos a estudiar, con carácter meramente ilustrativo, la gráfica de la primera derivada $f'(x)$ para constatar que no se anula en el intervalo, lo que garantiza, desde el punto de vista matemático, que la raíz es única.</p>

```python
fder(x) = f.diff(x)

plot(fder, (x, 0, 2))
```

<p>Podemos observar en la gráfica anterior que la derivada no se anula dentro del intervalo [0,2], por lo que la raíz es única.</p>

<p>Pasamos a realizar iteraciones del método de bisección. En la primera iteración, llamaremos</p>
<p>$a=0$,   $b=2$,  $\displaystyle x_1=\frac{a+b}{2}$</p>
<p>para evaluar a continuación $f(x_1)$.</p>

```python
a = 0; b = 2; x1 = (a + b)/2;
print('aprox.: ', x1.n(),' valor f:', f(x1).n())
```

<p>Como $f(x_1)f(b)<0$, tomaremos ahora</p>
<p>$a=x1$,  $b=b$,   $\displaystyle x_2=\frac{a+b}{2}$</p>

```python
a = x1; x2 = (a + b)/2;
print('aprox.: ', x2.n(),' valor f:', f(x2).n())
```

<p>Ahora ha resultado  $f(x_2)f(a)<0$, luego elegimos</p>
<p>$a=a$,  $b=x2$,  $\displaystyle x_3=\frac{a+b}{2}$.</p>

```python
b = x2; x3 = (a + b)/2;
print('aprox.: ', x3.n(),' valor f:', f(x3).n())
```

<p>De nuevo ha resultado  $f(x_3)f(a)<0$, luego elegimos</p>
<p>$a=a$,  $b=x3$,  $\displaystyle x_4=\frac{a+b}{2}$.</p>

```python
b = x3; x4 = (a + b)/2;
print('aprox.: ', x4.n(),' valor f:', f(x4).n())
```

<p>Podríamos seguir iterando el proceso, pero es interesante poder acotar el error cometido en el último paso. Para ello usamos la cota del <strong>error a posteriori</strong>, es decir,</p>
<p>$$\varepsilon_n\leq\frac{|f(x_n)|}{\min_{x\in[a,b]}|f'(x)|}$$</p>
<p>Necesitamos una cota inferior del valor mínimo del valor absoluto de la derivada dentro del intervalo. Vamos a ayudarnos de una representación gráfica.</p>

```python
fder(x) = f.diff(x)

plot(abs(fder), (x, 0, 2))
```

<p>El valor mínimo se alcanza en este caso en $x=0$. Hallamos el valor:</p>

```python
minder = abs(fder(0.))
minder
```

<p>Ahora podemos aplicar la fórmula del error a posteriori:</p>

```python
error = abs(f(x4))/minder; error.n()
```

<p>También podemos prever cuantas iteraciones serán necesarias como mínimo para garantizar un determinado número de cifras decimales exactas, utilizando la fórmula del <strong>error a priori </strong>para la sucesión de aproximaciones<strong> $\{x_1,x_2,\dots,x_n\}$:</strong> obtenidas en el método de bisección</p>
<p>$$\epsilon_n\leq \frac{b-a}{2^n}$$</p>
<p>Por ejemplo, calculemos el número de iteraciones necesarias para garantizar un error menor que $10^{-3}$, es decir, garantizar tres cifras decimales exactas.</p> Se garantizará esta precisión siempre que $\frac{b-a}{2^n}\leq 10^{-3}$. Se tiene: 
<p>$$ \frac{b-a}{2^n}\leq 10^{-3}\ \Leftrightarrow 2^n\ge 10^3(b-a)\Leftrightarrow\ n\ge \log_2(10^3(b-a))$$</p>

```python
a = 0; b = 2; cifras = 3

log(10^cifras*(b - a), 2).n()  ## Ojo! En SageMath la función log(x) representa el logaritmo neperiano o natural.
                               ## El logaritmo en base 2 de x se expresa como log(x, 2)
```

<p>Luego se precisan 11 iteraciones para garantizar, a priori, tres cifras decimales exactas.</p>

<p>A continuación diseñaremos un algoritmo simple donde cada iteración del método de bisección  es implementada utilizando un bucle <span style="color: #0000ff;">for</span>. Tomaremos como intervalo inicial  [0,2]. Estamos interesados en calcular 11 iteraciones del método y en cada paso proporcionaremos los valores de $x_i$ , de $f(x_i)$  y de la cota del error a priori:</p>

```python
iteraciones = 11  ; a = 0. ;  b = 2. ;  f(x) = 3*sin(x/2)-5*cos(x)  ## DATOS INTRODUCIDOS POR EL USUARIO Y MODIFICABLES

for i in range(iteraciones):                             ##  ESTE BLOQUE DE CÓDIGO NO HAY PORQUÉ MODIFICARLO
    xi = (a + b)/2
    print('x_', i + 1, ':', xi.n(),' f(x', i + 1, ')=', f(xi), ' error apriori:', ((b - a)/2).n())
    if f(a)*f(xi) < 0:
        b = xi
    else:
        a = xi
```

<p>En la última iteración tan solo podemos asegurar que tenemos 3 cifras decimales exactas, tras 11 iteraciones.</p>
<p>Podemos ver, para el último valor calculado, su cota del error a posteriori:</p>

```python
(abs(f(xi))/minder).n()   ## minder es el mínimo de la derivada, calculado anteriormente
```

<p>Finalmente proporcionamos una versión mejorada gráficamente del algoritmo de bisección que toma como base el algoritmo anterior. Para simplificar y automatizar el proceso, hemos creado la función</p>
<p>                            <span style="color: #0000ff;">m_biseccion(f, a, b, niter, digits)</span></p>
<p>donde f es la función asociada a la ecuación $f(x)=0$, considerada en el intervalo $[a,b]$, niter representa el número de iteraciones a realizar y digits es un parámetro opcional para indicar el número de cifras con el que expresar el resultado; por defecto se muestran 8 dígitos.</p>

```python
load('codigo_examinar_html.sage') # para definir show_html

def m_biseccion(f, a, b, iter, digitos=8):
    def N(x):
        return x.n(digits = digitos)
    codigo_latex = 'M. bisección para $%s =0$ en el intervalo $[%s,%s]$' % (latex(f(x)), a, b)
    L = abs(b-a)
    x0 = (a + b)/2
    v = (1,N(x0), 1, N(f(x0)), 1, N(L/2))
    valores = [ v ]
    for i in range(2, iter + 1):
        if f(a) * f(x0) < 0:
            b = x0
        else:
            a = x0
        x0 = (a + b)/2
        valores.append( (i, N(x0), i, N(f(x0)), i , N(L/2^(i))) )
    s = r' \\ x_{%s} & =  %s  & \quad  f(x_{%s}) & =  %s  & \varepsilon_{%s} &<  %s'
    contenido = "".join([s%v for v in valores])
    tabla = r'\begin{align*} %s \end{align*} '%contenido
    show_html('<font color="blue"> <center> %s   %s </center>' % (codigo_latex, tabla))
    return x0.n(digits=digitos)
```

<p>Ponemos a prueba el nuevo algoritmo con el ejemplo anterior, mostrando 6 dígitos decimales:</p>

```python
f(x) = 3*sin(x/2) - 5*cos(x)

m_biseccion(f, 0, 2, 11, 6)
```

<h4>Aplicación a la resolución de ecuaciones $g(x)=h(x)$</h4>
<p>El método de bisección también puede ser utilizado para determinar el punto de corte de dos gráficas de funciones $g(x)=h(x)$ con tan solo considerar la nueva función, diferencia de ambas, $f(x)=g(x)-h(x)$. Veamos un ejemplo.</p>

```python
a = 4; b = 5;
g(x) = 2 + log(x)
h(x) = exp(x/2) - 5

show(
  plot(g(x), a, b, color = 'blue')
  + plot(h(x), a, b, color = 'red')
)
```

```python
m_biseccion(g - h, 4, 5, 20)
```

<p><a name="puntofijo"></a></p>
<h3>Método del punto fijo</h3>
<p>Recordemos que el método del punto fijo pretende resolver ecuaciones de la forma $$x=g(x).$$</p>
<p>El <strong>teorema del punto fijo</strong> garantiza que si $g(x)$ es continua en un intervalo $[a,b]$, de forma que $g([a,b])\subset [a,b]$, y su derivada verifica que $|g'(x)|\leq k<1$ en el interior del intervalo $(a,b)$, entonces la ecuación $x=g(x)$ tiene una única solución (punto fijo de $g(x)$). Además, la sucesión de valores $x_{n+1}=g(x_n)$ converge a la solución para cualquier valor inicial $x_0$ tomado en el intervalo $[a,b]$.</p>
<p>Ilustraremos este resultado con el siguiente ejemplo dentro del intervalo $[0,4]$:</p>
<p>$$x=\frac{x+2}{x+1}$$</p>

```python
a = 0; b = 4;
g(x) = (x + 2)/(x + 1)
h(x) = x

plot(g(x), a, b, color = 'blue') + plot(h(x), a, b, color = 'red')  + rectangulo((a, a), (b, b))
```

<p>Podemos comprobar como la gráfica de $g(x)$, de color azul, queda dentro del cuadrado rojo, lo que garantiza la condición $g([a,b])\subset [a,b]$. Estudiemos ahora la derivada de $g(x)$.</p>

```python
gder = g.diff(x); show(simplify(factor( gder(x)  )))
## la combinación de las funciones simplify(factor(  )) simplifica las expresiones matemáticas
```

```python
show(
plot(abs(gder(x)), a, b, color = 'blue') +
rectangulo((a, 0), (b, 1))
)
```

<p>La función $|g'(x)|$ toma valores entre 0 y 1, pero no existe ningún valor k verificando $|g'(x)|\leq k<1$ pues se alcanza el valor 1 en $x=0$. Por tanto, vamos a restringir el intervalo de búsqueda a otro menor que contenga el punto de corte (punto fijo), por ejemplo, el intervalo $[1,2]$.</p>

```python
a = 1; b = 2;
g(x) = (x + 2)/(x + 1)
h(x) = x

show(
  plot(g(x), a, b, color = 'blue')
  + plot(h(x) ,a, b, color = 'red')
  + rectangulo((a, a), (b, b))
)
```

```python
show(
plot(abs(gder(x)), a, b, color = 'blue') +
rectangulo((a, 0), (b, 1))
)
```

<p>En este caso, la cota superior para $|g'(x)|$ se alcanza en $x=1$. Veamos su valor:</p>

```python
cotasup = abs(gder(1.))
cotasup
```

<p>Por tanto, ahora, en el intervalo $[1,2]$ sí se verifican las condiciones del teorema del punto fijo, por lo que ya tenemos garantizada la convergencia de la sucesión $x_{n+1}=g(x_n)$. Empezaremos, por ejemplo, por $x_0=1$.</p>

<p><a name="metodopuntofijo"></a></p>
<h3>Algoritmo del método de punto fijo</h3>
<p>Una vez garantizada la convergencia del método de punto fijo, aplicado a la ecuación $x=g(x)$  dentro de cierto intervalo $[a,b]$, basta elegir un valor inicial $x_0\in [a,b]$ y hallar valores de forma iterada con $x_{n+1}=g(x_n)$.  </p>

```python
x0 = 1.
x1 = g(x0)
x1
```

```python
x2 = g(x1)
x2
```

```python
x3 = g(x2)
x3
```

```python
x4 = g(x3)
x4
```

<p>Así podemos continuar indefinidamente. No obstante, podemos analizar el <strong>error a priori </strong>para determinar el número mínimo de iteraciones necesarias para garantizar un determinado número de cifras decimales exactas. Recordemos que la fórmula es:</p>
<p>$$\epsilon_n\leq k^n\cdot \max(x_0-a,b-x_0)$$</p>
<p>para un valor $k$ verificando $|g'(x)|\leq k<1$.</p>
<p>Veamos cuántas iteraciones son necesarias para garantizar 3 cifras decimales exactas:</p>

$$\begin{align*}
k^n\cdot \max(x_0-a,b-x_0)\leq 10^{-3}\\ 
&\Leftrightarrow k^n\leq \frac{10^{-3}}{\max(x_0-a,b-x_0)} \\
&\Leftrightarrow\ n\cdot \log(k)\leq \log\left(\frac{10^{-3}}{\max(x_0-a,b-x_0)}\right) \\
&\Leftrightarrow n\ge \log\left(\frac{10^{-3}}{\max(x_0-a,b-x_0)}\right)/\log(k)
\end{align*}$$

```python
cotasup_k = 0.25; x0 = 1.; a = 1; b = 2; cifras = 3

### no hace falta modificar el siguiente código:
iter = log( 10^(-cifras) / max(x0 - a, b - x0))/ log(cotasup_k);
iter
```

<p>Luego son precisas 5 iteraciones.</p>

```python
x5 = g(x4)
x5
```

<p>También podemos determinar el error<strong> a posteriori</strong> para la última aproximación, utilizado la fórmula:</p>
<p>$$\varepsilon_n\leq\frac{|f(x_n)|}{\min_{x\in[a,b]}|f'(x)|}$$</p>
<p>donde es importante recordar que $f(x)=x-g(x)$, por lo que se requiere acotar inferiormente el valor absoluto  de la derivada de $f$ dentro del intervalo.</p>

```python
f(x) = x - g(x)
derf = f.diff()
plot(abs(derf), a, b)
```

<p>En vista de la gráfica anterior, la cota inferior del valor absoluto de la derivada se obtiene en $x=2.$</p>

```python
cotainf = abs(derf( x = 2.)); cotainf
```

<p>Y la cota del error a posteriori es:</p>

```python
abs(f(x = x5))/cotainf
```

<p>Por tanto $x_5$ es una aproximación a la solución con al menos 4 cifras decimales exactas.</p>

```python
x5
```

<p>A continuación diseñaremos un algoritmo simple donde cada iteración del método de punto fijo es implementada utilizando un bucle <span style="color: #0000ff;">for</span>. Tomaremos como intervalo inicial  [1,2]. Estamos interesados en calcular 5 iteraciones del método:</p>

```python
iteraciones = 5; x0 = 1.

xi = x0                                             ## NO ES PRECISO MODIFICAR ESTE CÓDIGO
for i in range(iteraciones):                                                 
    xi = g(xi)
    print('iter.:', i + 1, ' valor:', xi)
```

<p>Finalmente proporcionamos una versión mejorada gráficamente del método de punto fijo que toma como base el algoritmo anterior. Para simplificar y automatizar el proceso, hemos creado la función</p>
<p>                            <span style="color: #0000ff;">m_punto_fijo($g$, $x_0$, niter, digits)</span></p>
<p>donde $g$ es la función asociada a la ecuación $x=g(x)$, el valor $x_0$ es la aproximacion inicial, niter representa el número de iteraciones a realizar y digits es un parámetro opcional para indicar el número de cifras con el que expresar el resultado; por defecto se muestran 8 dígitos.</p>

```python
load('codigo_examinar_html.sage') # para definir show_html

def m_punto_fijo(f, x0, niter, digitos=8):
    def N(x):
        return x.n(digits=digitos)
    valores = [ (0, N(x0)) ]
    xi = x0
    for i in range(1, niter+1):
        xi = N(f(xi))
        valores.append((i, xi))
    contenido = ""
    s = r' \\  x_{%s} & = & %s  & \quad  x_{%s} & = & %s'
    if is_even(niter):
        contenido += r' \\  x_{%s} & = & %s  &  & & ' % (0, N(x0))
        contenido += "".join([s%(i1, x1 ,i2, x2) for ((i1, x1), (i2, x2)) in zip(valores[1:],valores[1+ niter/2:])])
    else:
        contenido += "".join([s%(i1, x1 ,i2, x2) for ((i1, x1), (i2, x2)) in zip(valores,valores[(niter+1)/2:])])
    tabla = r'\begin{array}{lcrlcr} %s \end{array} '%contenido
    codigo_latex =r' Iteraciones de $x_{n+1}=g(x_n)$, para $\displaystyle g(x)=%s$  $$ %s $$' % (latex(f(x)),tabla)
    show_html('<font color="blue"> <center> %s  </center>' % codigo_latex)
    return xi
```

```python
g(x) = (x + 2)/(x + 1)
```

```python
m_punto_fijo(g, 1.5, 6, 16)
```
