```python
load('codigo_examinar_html.sage') # para definir show_html

def m_biseccion(f, a, b, iter, digitos=8):
    def N(x):
        return x.n(digits = digitos)
    codigo_latex = 'M. bisección para $%s =0$ en el intervalo $[%s,%s]$' % (latex(f(x)), a, b)
    L = abs(b-a)
    x0 = (a + b)/2
    v = (1,N(x0), 1, N(f(x0)), 1, N(L/2))
    valores = [ v ]
    for i in range(2, iter + 1):
        if f(a) * f(x0) < 0:
            b = x0
        else:
            a = x0
        x0 = (a + b)/2
        valores.append( (i, N(x0), i, N(f(x0)), i , N(L/2^(i))) )
    s = r' \\ x_{%s} & =  %s  & \quad  f(x_{%s}) & =  %s  & \varepsilon_{%s} &<  %s'
    contenido = "".join([s%v for v in valores])
    tabla = r'\begin{align*} %s \end{align*} '%contenido
    show_html('<font color="blue"> <center> %s   %s </center>' % (codigo_latex, tabla))
    return x0.n(digits=digitos)
```

```python
load('codigo_examinar_html.sage') # para definir show_html

def m_punto_fijo(f, x0, niter, digitos=8):
    def N(x):
        return x.n(digits=digitos)
    valores = [ (0, N(x0)) ]
    xi = x0
    for i in range(1, niter+1):
        xi = N(f(xi))
        valores.append((i, xi))
    contenido = ""
    s = r' \\  x_{%s} & = & %s  & \quad  x_{%s} & = & %s'
    if is_even(niter):
        contenido += r' \\  x_{%s} & = & %s  &  & & ' % (0, N(x0))
        contenido += "".join([s%(i1, x1 ,i2, x2) for ((i1, x1), (i2, x2)) in zip(valores[1:],valores[1+ niter/2:])])
    else:
        contenido += "".join([s%(i1, x1 ,i2, x2) for ((i1, x1), (i2, x2)) in zip(valores,valores[(niter+1)/2:])])
    tabla = r'\begin{array}{lcrlcr} %s \end{array} '%contenido
    codigo_latex =r' Iteraciones de $x_{n+1}=g(x_n)$, para $\displaystyle g(x)=%s$  $$ %s $$' % (latex(f(x)),tabla)
    show_html('<font color="blue"> <center> %s  </center>' % codigo_latex)
    return xi
```
